```python
speed_mean = playoff["release_speed"].mean()

speed_mean
```

```python
se = np.std(playoff["release_speed"]) / np .sqrt(len(playoff["release_speed"]))
```

```python
velo_mean_dist = stats.norm(speed_mean, se)


lower, upper = velo_mean_dist.ppf(0.025),velo_mean_dist.ppf(0.975)

print(f'Lower 95% CI: {lower:2.2f}')
print(f'Upper 95% CI: {upper:2.2f}')

```
